```python
from tkinter import*
import tkinter.messagebox
```

```python
class beginner:
    
    def __init__(self, root):
        self.root = root
        self.root.title("Addition")
        self.root.geometry("400x200")
        
        frame1= Frame(self.root)
        frame1.grid()
        
        frame2= Frame(frame1, width=400, height=100)
        frame2.grid(row=0, column=0)
         
        frame3= Frame(frame1, width=400, height=100)
        frame3.grid(row=1, column=0)
        
        FirstNum=StringVar()
        SecondNum=StringVar()
        Result=StringVar()
        
        self.lblFirstNum = Label(frame2, text='Enter first number')
        self.lblFirstNum.grid(row=0, column=0)
        self.textFirstNum = Entry(frame2, textvariable=FirstNum)
        self.textFirstNum.grid(row=0, column=1)
        
        self.lblSecondNum = Label(frame2, text='Enter Second number')
        self.lblSecondNum.grid(row=1, column=0)
        self.textSecondNum = Entry(frame2, textvariable=SecondNum)
        self.textSecondNum.grid(row=1, column=1)
        
        self.lblResult = Label(frame2, text='Show Result')
        self.lblResult.grid(row=2, column=0)
        self.textResult = Entry(frame2, textvariable=Result)
        self.textResult.grid(row=2, column=1)
        
        
        def add():
            f = float(FirstNum.get())
            s = float(SecondNum.get())
            r = f + s
            Result.set(r)
            
        def reset():
            FirstNum.set("")
            SecondNum.set("")
            Result.set("")
            
        def exit():
            root.destroy()
            
        self.btnAdd = Button(frame3, text="Add Data", width=12, height=2, command=add)
        self.btnAdd.grid(row=0, column=0)
        
        self.btnReset= Button(frame3, text="Reset", width=12, height=2, command=reset)
        self.btnReset.grid(row=0, column=1)
        
        self.btnExit= Button(frame3, text="Exit", width=12, height=2, command=exit)
        self.btnExit.grid(row=0, column=2)
    
    
        
if __name__=='__main__':
    root=Tk()
    app = beginner(root)
    root.mainloop()

    
    
    
    
```
